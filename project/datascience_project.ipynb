# Streaming Viewers Analysis


Streaming Membership Services get many clients over the world wanting to watch their own shows at their own time instead of watching cable which is more expensive and you also watch ads with how expensive it is, but different people get different tiers of their membership and also multiple people watching different genres and also on different languages depending on where they live.

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
df = pd.read_csv('streaming_viewership_data.csv')
df['People'] = 1
df['Subscription_Status'] = df['Subscription_Status'].replace({'Premium':1,'Free':0})
df.head()
```

### #1 How many people by gender watch per genre?



```python
dos = df.groupby(['Gender', 'Genre'])['People'].sum()
dos.plot(kind='bar', color=('#660066','#FF4040'))
plt.title('People per genre by gender')
plt.ylabel('Total amount of people')
plt.xlabel('Gender and Gender')
```

As seen by the graph by the difference with Female and Male both genders have different interest as shown with Females having Thrillers as their top most watch genre and Males having Documentaries as their top one of watching.

### #2 What devices are used for streaming by gender?

```python
tres = df.groupby(['Device_Type','Gender'])['People'].sum()
tres.plot(kind='bar', color=('#088DA5','#8A2BE2'))
plt.title('Total Devices per Gender')
plt.ylabel('Total People')
plt.xlabel('Device and Gender')
```

It is interesting to note the difference in values between both gender in the graph is knowing where between gender both preffer to watch their shows different for example they could be watching from their phones or laptops but as shown in the graph between male and female 

### #3 How many people are subscribed per gender?

```python
cuatro = df.groupby(['Gender'])['Subscription_Status'].sum()
cuatro.plot(kind='bar',color=('#e5a3ad','#3399FF'))
plt.title('Total Subscription per Gender')
plt.ylabel('Subscribed')
plt.xlabel('Gender')
```

As seen in the graph above we can see that between both genders it is very similar the rate of subscriptions but between them the males are the ones who are the most interested on getting the premium subscription to get rid of ads and to get more shows compared to Female. 

### #4 What is the quality of video usually watched per genre?

```python
cinco = df.groupby(['Playback_Quality','Genre'])['People'].sum()
cinco.plot(kind='bar', color=('#FFFF66','#8B0000'))
plt.title('Average Quality per Genre')
plt.ylabel('Video quality and Genre')
plt.xlabel('Total People')
```

Comparing the different data series on the graph, we can see that will all the genre and video quality the one with the most viewers would be documentaries with HD video quality with Thriller as second with the worst quality SD.
