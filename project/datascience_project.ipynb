# Streaming Viewers Analysis


Streaming Membership Services get many clients all over the world wanting to watch their own shows on their own time instead of watching cable which is more expensive and also having to watch ads with how expensive it is, but different people get different tiers of their membership, and also multiple people watching different genres and also on different languages depending on where they live.

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
df = pd.read_csv('streaming_viewership_data.csv')
df['People'] = 1
df['Subscription_Status'] = df['Subscription_Status'].replace({'Premium':1,'Free':0})
df.head()
```

### #1 What is are the most popular genre by gender?



```python
dos = df.groupby(['Gender', 'Genre'])['People'].sum()
dos.plot(kind='bar', color=('#660066','#FF4040'))
plt.title('People per Genre and Gender')
plt.ylabel('Total amount of people')
plt.xlabel('Gender and Genre')
```

As seen by the graph the difference between Females and Males both have different interests as shown with Females having Thrillers as their top most watched genre and Males having Documentaries as their top one to watch. It is also notisable by having their least watched for example having Drama as the least preferred by Females and for Males having both Comedy and Thriller as their least watched.

### #2 What devices are used for streaming by gender?

```python
tres = df.groupby(['Device_Type','Gender'])['People'].sum()
tres.plot(kind='bar', color=('#088DA5','#8A2BE2'))
plt.title('Total Devices per Gender')
plt.ylabel('Total People')
plt.xlabel('Device and Gender')
```

It is interesting to note the difference in values between both genders in the graph is knowing where each gender prefers to watch their shows. As shown in the graph they use different devices like watching from their phones or laptops but this could also be different depending on where the person is currently for example they could be out of home and they wanna watch something so they use their phone or they are at home so they use their pc or smart tv.

### #3 How many people are subscribed per gender?

```python
cuatro = df.groupby(['Gender'])['Subscription_Status'].sum()
cuatro.plot(kind='bar',color=('#e5a3ad','#3399FF'))
plt.title('Total Subscription per Gender')
plt.ylabel('Total Subscribed')
plt.xlabel('Gender')
```

As seen in the graph above we can see that between both genders it is very similar the rate of subscriptions but between them, the males are the ones who are the most interested in getting the premium subscription to get rid of ads and to get more shows compared to females which this can also be dependant on their needs to watch and their financial needs. 

### #4 What is the quality of video usually watched per genre?

```python
cinco = df.groupby(['Playback_Quality','Genre'])['People'].sum()
cinco.plot(kind='bar', color=('#FFFF66','#8B0000'))
plt.title('Average Quality per Genre')
plt.ylabel('Total People')
plt.xlabel('Video Quality and Genre')
```

Comparing the different data series on the graph, we can see all the genres and video quality that viewers have watched combined together over the time of the data recorded. Among all the genres watched at their video qualities would be documentaries with HD video quality with Thriller as second with the worst quality SD and these results are also based on the graph in question one because it is also dependent on the number of people watching one genre and the other.
